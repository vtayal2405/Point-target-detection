```python
import numpy as np
import cv2
from matplotlib import pyplot as plt
from scipy.ndimage import gaussian_filter
from skimage import io
```

```python
img = cv2.imread("Pictures/AIRPLANE_000231.png", cv2.IMREAD_GRAYSCALE)
img = img.astype(np.float32)
plt.imshow(img, cmap='gray')
plt.title("Original IR Image")
plt.show()
```

```python
def compute_point_spread_indicator(image):
    M, N = image.shape
    padded = np.pad(image, 1, mode='reflect')
    p_map = np.zeros_like(image)

    for i in range(1, M+1):
        for j in range(1, N+1):
            C = padded[i, j]
            Fm = [padded[i-1,j], padded[i+1,j], padded[i,j-1], padded[i,j+1]]
            Fn = [padded[i-1,j-1], padded[i-1,j+1], padded[i+1,j-1], padded[i+1,j+1]]

            IM = np.mean(Fm)
            IN = np.mean(Fn)
            if C > 1e-5 and IM > 1e-5 and IN > 1e-5:
                p = (np.log(C) - np.log(IM)) / (np.log(C) - np.log(IN))
                p_map[i-1, j-1] = p
    return p_map
```

```python
def target_protecting_filter(image, p_map, low=0.45, high=0.55):
    filtered = cv2.medianBlur(image.astype(np.uint8), 3).astype(np.float32)
    output = np.where((p_map >= low) & (p_map <= high), image, filtered)
    return output

p_map = compute_point_spread_indicator(img)
filtered_img = target_protecting_filter(img, p_map)

plt.imshow(filtered_img, cmap='gray')
plt.title("After Filtering (Target-Protecting Filter)")
plt.show()
```

```python
def point_spread_local_contrast(image):
    M, N = image.shape
    padded = np.pad(image, 1, mode='reflect')
    R = np.zeros_like(image)

    for i in range(1, M+1):
        for j in range(1, N+1):
            I0 = padded[i, j]
            Ij = [padded[i-1,j], padded[i+1,j], padded[i,j-1], padded[i,j+1]]
            Ii = [padded[i-1,j-1], padded[i-1,j], padded[i-1,j+1],
                  padded[i,j-1], padded[i,j+1],
                  padded[i+1,j-1], padded[i+1,j], padded[i+1,j+1]]

            if min(Ij) > 1e-5 and min(Ii) > 1e-5:
                L = min([(I0**2)/ii for ii in Ii])
                R[i-1, j-1] = (min(Ij) / max(Ij)) * L
    return R
```

```python
def high_boost_filter(M, kernel_size=9):
    MF = gaussian_filter(M, sigma=kernel_size/6.0)
    B = np.maximum(M - MF, 1)
    E = M * B
    return E

M = point_spread_local_contrast(filtered_img)
enhanced = high_boost_filter(M)
plt.imshow(enhanced, cmap='hot')
plt.title("Enhanced Image (PSLCM + High-Boost)")
plt.show()
```

```python
def adaptive_segmentation(E):
    V = np.sort(E.ravel())[::-1]
    S = np.unique(V)
    len_S = len(S)

    for i in range(min(10, len_S-2), 1, -1):
        r = (S[i] - S[i+1]) / (S[i+1] - S[i+2] + 1e-6)
        if r > 5:
            T = S[i]
            break
        elif r < 0.2:
            T = S[i+1]
            break
        else:
            T = V[9]
    return T

T = adaptive_segmentation(enhanced)
binary = (enhanced >= T).astype(np.uint8) * 255

plt.imshow(binary, cmap='gray')
plt.title(f"Segmentation Result (T = {T:.2f})")
plt.show()
```
