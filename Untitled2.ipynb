```python
import cv2
import numpy as np
import matplotlib.pyplot as plt





```

```python
img = cv2.imread("Pictures/AIRPLANE_000231.png", cv2.IMREAD_GRAYSCALE).astype(np.float32)

plt.figure(figsize=(12, 4))
plt.plot(1, 3, 1)
plt.imshow(img, cmap='gray')
plt.title("Original IR Image")
```

```python
blurred = cv2.GaussianBlur(img, (21, 21), 0)

foreground = img - blurred

norm = cv2.normalize(foreground, None, 0, 255, cv2.NORM_MINMAX)
norm = norm.astype(np.uint8)

plt.plot(1, 3, 2)
plt.imshow(norm, cmap='gray')
plt.title("After Background Subtraction")
```
